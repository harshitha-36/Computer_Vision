```python
#21BAI1029
#Import libraries and color image.
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/mvphoto.jpeg")
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#a) Image Negative

gray_negative = abs(255-img_gray)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('gray_img')
plt.subplot(1,2,2)
plt.imshow(gray_negative, cmap='gray')
plt.title('Negative')
```

```python
#b) Log-Transform
log1=2*np.log(img_gray +1)
log2=34*np.log(img_gray +1)
```

```python
plt.figure(figsize=(9, 3))
plt.subplot(1,3,1)
plt.imshow(img_gray,cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(np.uint8(log1), cmap='gray')
plt.title('Log 1')
plt.subplot(1,3,3)
plt.imshow(np.uint8(log2),cmap='gray')
plt.title('Log 2')
```

```python
#c) Gamma Correction
gamma=1.89
m, n = img_gray.shape
gamma1=np.zeros([m,n])
for i in range(m):
  for j in range(n):
    gamma1[i,j]=img_gray[i,j]**gamma
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(np.uint8(gamma1), cmap='gray')
plt.title('Gamma')
```

```python
#Inference
plt.figure(figsize=(10, 5))
plt.subplot(2,2,1)
plt.hist(img_gray.ravel(), 256,[0,255],density=1)
plt.title('Original')
plt.subplot(2,2,2)
plt.hist(log1.ravel(), 256,[0,255],density=1)
plt.title('Log')
plt.subplot(2,2,3)
plt.hist(gamma1.ravel(), 256,[0,255],density=1)
plt.title('Gamma')
plt.subplot(2,2,4)
plt.hist(gray_negative.ravel(), 256,[0,255],density=1)
plt.title('Negative')
plt.show()
```

```python
#Task 2) Perform an operation on the image such that gray values < threshold is
#zero and > threshold is 255 (Image Binarization).
```

```python
(m,n)=img_gray.shape
img1=np.zeros([m,n])
img1.shape
```

```python
t=76
for i in range(1346):
  for j in range(2048):
    if img_gray[i,j]<t:
      img1[i,j]=0
    else:
      img1[i,j]=67
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img1, cmap='gray')
plt.title('Binarization')
```

```python
#Task 3) Enhance the input gray image using contrast stretching method.
#Observe the output and also try to plot the histogram of the input and the processed image.
```

```python
contrast= img_gray.copy()
xp= [0,50, 128, 192, 255]
fp = [0,25,200,225,255]
x=np.arange(256)
table = np.interp(x,xp,fp).astype('uint8')
contrast =cv2.LUT(contrast, table)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(contrast, cmap='gray')
plt.title('Contrast')
```

```python
#Histogram
plt.figure(figsize=(8, 4))
plt.subplot(1,2,1)
plt.hist(img_gray.ravel(), 256,[0,255],density=1)
plt.title('Original')
plt.subplot(1,2,2)
plt.hist(contrast.ravel(), 256,[0,255],density=1)
plt.title('Constrast')
plt.show()
```
