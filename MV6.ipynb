```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def region_growing(img, seed):
    height, width = img.shape
    segmented = np.zeros_like(img, dtype=np.uint8)
    threshold = 80
    neighbors = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    stack = [seed]
    seed_intensity = img[seed]

    while stack:
        current_seed = stack.pop()
        if 0 <= current_seed[0] < height and 0 <= current_seed[1] < width:
            intensity_diff = abs(int(img[current_seed]) - int(seed_intensity))
            if intensity_diff < threshold and segmented[current_seed] == 0:
                segmented[current_seed] = 255
                for neighbor in neighbors:
                    new_seed = (current_seed[0] + neighbor[0], current_seed[1] + neighbor[1])
                    stack.append(new_seed)
    return segmented
```

```python
image = cv2.imread('/content/coin.jpg', cv2.IMREAD_GRAYSCALE)
seed_point1 = (100, 140)
seed_point2 = (100, 240)
seed_point3 = (100, 340)
seed_point4 = (180, 120)
seed_point5 = (180, 240)
seed_point6 = (195, 361)
seed_point7 = (280, 330)
seed_point8 = (280, 240)
segmented_image1 = region_growing(image, seed_point1)
segmented_image2 = region_growing(image, seed_point2)
segmented_image3 = region_growing(image, seed_point3)
segmented_image4 = region_growing(image, seed_point4)
segmented_image5 = region_growing(image, seed_point5)
segmented_image6 = region_growing(image, seed_point6)
segmented_image7 = region_growing(image, seed_point7)
segmented_image8 = region_growing(image, seed_point8)

```

```python
plt.figure(figsize=(20,14))
plt.subplot(3,3,1)
plt.imshow(image,'gray')
plt.title("Original Image")

plt.subplot(3,3,2)
plt.imshow(segmented_image1,'gray')
plt.title("Region Segmented Image")

plt.subplot(3,3,3)
plt.imshow(segmented_image2,'gray')
plt.title("Region Segmented Image")

plt.subplot(3,3,4)
plt.imshow(segmented_image3,'gray')
plt.title("Region Segmented Image")
plt.subplot(3,3,5)
plt.imshow(segmented_image4,'gray')
plt.title("Region Segmented Image")

plt.subplot(3,3,6)
plt.imshow(segmented_image5,'gray')
plt.title("Region Segmented Image")
plt.subplot(3,3,7)
plt.imshow(segmented_image6,'gray')
plt.title("Region Segmented Image")

plt.subplot(3,3,8)
plt.imshow(segmented_image8,'gray')
plt.title("Region Segmented Image")
plt.subplot(3,3,9)
plt.imshow(segmented_image7,'gray')
plt.title("Region Segmented Image")
```

```python
def add_noise(image, mean=0, sigma=100):
    row, col = image.shape
    gauss = np.random.normal(mean, sigma, (row, col))
    noisy = np.clip(image + gauss, 0, 255)
    return noisy.astype(np.uint8)

noisy_image = add_noise(image)

plt.subplot(1, 2, 1)
plt.imshow(image, 'gray')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(noisy_image, 'gray')
plt.title('Noisy Image')
plt.show()
```
