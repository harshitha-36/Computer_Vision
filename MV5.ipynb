```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/Shinchan.jpg")
plt.imshow(img)
plt.axis("off")
```

```python
img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
horizontal = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]])
vertical = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]])
diagonal = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]])
horizontal1 = cv2.filter2D(img_gray, -1, horizontal)
vertical1 = cv2.filter2D(img_gray, -1, vertical)
diagonal1 = cv2.filter2D(img_gray, -1, diagonal)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(horizontal1, cmap='gray')
plt.title('Horizontal')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(vertical1, cmap='gray')
plt.title('Vertical')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(diagonal1, cmap='gray')
plt.title('Diagonal')
plt.axis('off')
```

```python
sobel_x = np.array([[-1,-2,-1], [0,0,0],[1,2,1]])
sobel_y = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
image_x = cv2.filter2D(img_gray, -1, sobel_x)
image_y = cv2.filter2D(img_gray, -1, sobel_y)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image_x, cmap='gray')
plt.title('sobel_x')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(image_y, cmap='gray')
plt.title('sobel_y')
plt.axis('off')
```

```python
perwit_x = np.array([[-1,-1,-1], [0,0,0],[1,1,1]])
perwit_y = np.array([[-1,0,1], [-1,0,1],[-1,0,1]])
image1_x = cv2.filter2D(img_gray, -1, perwit_x)
image2_y = cv2.filter2D(img_gray, -1, perwit_y)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image1_x, cmap='gray')
plt.title('Perwit_x')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(image2_y, cmap='gray')
plt.title('Perwit_y')
plt.axis('off')
```

```python
roberts_x = np.array([[1,0], [0,-1]])
roberts_y = np.array([[0,-1], [1,0]])
robert1 = cv2.filter2D(img_gray, -1, roberts_x)
robert2 = cv2.filter2D(img_gray, -1, roberts_y)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(robert1, cmap='gray')
plt.title('Roberts_x')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(robert2, cmap='gray')
plt.title('Roberts_y')
plt.axis('off')
```

```python
img1= cv2.imread("/content/noiseimage_salt_pepper.jpg")
plt.imshow(img1)
plt.axis("off")
```

```python
image_gray= cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
plt.hist(image_gray.ravel(), 256, [0,255], density=1)
plt.title('histogram')
```

```python
median = cv2.medianBlur(image_gray, 5)
plt.imshow(median, cmap= 'gray')
```

```python
plt.hist(median.ravel(), 256, [0,255], density=1)
plt.title('histogram')
```

```python
[m,n]=median.shape
img1=np.zeros([m,n])
img1.shape
```

```python
t= 197
for i in range(459):
  for j in range(458):
    if median[i,j]<t:
      img1[i,j]= 255
    else:
      img1[i,j]=0
```

```python
plt.imshow(img1, cmap='gray')
```

```python
sobel_x = np.array([[-1,-2,-1], [0,0,0],[1,2,1]])
sobel_y = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
image_x = cv2.filter2D(img1, -1, sobel_x)
image_y = cv2.filter2D(img1, -1, sobel_y)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image_x, cmap='gray')
plt.title('sobel_x')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(image_y, cmap='gray')
plt.title('sobel_y')
plt.axis('off')
```

```python
sobel_x = np.array([[-1,-2,-1], [0,0,0],[1,2,1]])
sobel_y = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
image_x = cv2.filter2D(median, -1, sobel_x)
image_y = cv2.filter2D(median, -1, sobel_y)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image_x, cmap='gray')
plt.title('sobel_x')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(image_y, cmap='gray')
plt.title('sobel_y')
plt.axis('off')
```

```python
sobelx = cv2.Sobel(median, cv2.CV_64F, 1, 0, ksize=7)
sobely = cv2.Sobel(median, cv2.CV_64F, 0, 1, ksize=7)
edges = np.sqrt(sobelx*2 + sobely*2)
plt.subplot(1,3,1)
plt.imshow(median,cmap='gray')
plt.title("GRAY")
plt.subplot(1,3,2)
plt.imshow(median ,cmap='gray')
plt.title("filtered")
plt.subplot(1,3,3)
plt.imshow(edges,cmap='gray')
plt.title("SOBEL")
```
