```python
#21BAI1029
#Import color image and libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("/content/cells.jpg")
plt.imshow(img)
plt.axis("off")
```

**Contrast Enhancement using Histogram equalization**

```python
img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
equal = cv2.equalizeHist(img_gray)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(equal, cmap='gray')
plt.axis('off')
plt.title('Equalization')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(img_gray.ravel(),256,[0,255],density=1)
plt.title("Original")
plt.subplot(1,2,2)
plt.hist(equal.ravel(),256,[0,255],density=1)
plt.title("Equalization")
```

**image smoothening based on ideal filters**

```python
noise_img=cv2.imread("/content/noisy_img.jpg")
plt.imshow(noise_img, cmap='gray')
plt.axis("off")
```

```python
kernel_size=3
kernel_size1=7
kernel_size2=10
kernel=np.ones((kernel_size, kernel_size), np.float32)/20
kernel1=np.ones((kernel_size, kernel_size1), np.float32)/50
kernel2=np.ones((kernel_size, kernel_size2), np.float32)/80
```

```python
print(kernel)
```

```python
print(kernel2)
```

```python
print(kernel1)
```

```python
smooth=cv2.filter2D(noise_img,-1,kernel)
smooth1=cv2.filter2D(noise_img,-1,kernel1)
smooth2=cv2.filter2D(noise_img,-1,kernel2)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(noise_img, cmap='gray')
plt.axis('off')
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(smooth, cmap='gray')
plt.axis('off')
plt.title('Smooth-1')
plt.subplot(2,2,3)
plt.imshow(smooth1, cmap='gray')
plt.axis('off')
plt.title('Smooth-2')
plt.subplot(2,2,4)
plt.imshow(smooth2, cmap='gray')
plt.axis('off')
plt.title('Smooth-3')
```

```python
blur=cv2.GaussianBlur(noise_img,(15,15),0)
blur1=cv2.GaussianBlur(noise_img,(21,21),0)
blur2=cv2.GaussianBlur(noise_img,(31,31),0)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(noise_img, cmap='gray')
plt.axis('off')
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(blur, cmap='gray')
plt.axis('off')
plt.title('Gaussian_blur-1')
plt.subplot(2,2,3)
plt.imshow(blur1, cmap='gray')
plt.axis('off')
plt.title('Gaussian_blur-2')
plt.subplot(2,2,4)
plt.imshow(blur2, cmap='gray')
plt.axis('off')
plt.title('Gaussian_blur-3')
```

**Image Sharpening**

```python
sharpen_filter = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
sharpen_filter1 = np.array([[2,7,-1], [0,3,-6], [2,-4,-1]])
sharpen_filter2 = np.array([[-9,2,-6], [-4,0,9], [7,0,-1]])
img_sharp=cv2.filter2D(noise_img,-1,sharpen_filter)
img_sharp1=cv2.filter2D(noise_img,-1,sharpen_filter1)
img_sharp2=cv2.filter2D(noise_img,-1,sharpen_filter2)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(noise_img, cmap='gray')
plt.axis('off')
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(img_sharp, cmap='gray')
plt.axis('off')
plt.title('Sharp-1')
plt.subplot(2,2,3)
plt.imshow(img_sharp1, cmap='gray')
plt.axis('off')
plt.title('Sharp-2')
plt.subplot(2,2,4)
plt.imshow(img_sharp2, cmap='gray')
plt.axis('off')
plt.title('Sharp-3')
```
